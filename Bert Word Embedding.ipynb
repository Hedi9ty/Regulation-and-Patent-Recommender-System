```python
# 載入繁體版Jieba
analyzer = zh_split('dictionary/dict.txt.big')
# 載入自定義字典
analyzer.add_dictionary('dictionary/edu_dict.txt')
analyzer.add_dictionary('dictionary/company_dict.txt')
analyzer.add_dictionary('dictionary/geo_dict.txt')
analyzer.add_dictionary('dictionary/law_dict.txt')
analyzer.add_dictionary('dictionary/FCS_dict.txt')
# 斷詞
analyzer.split(df.技術領域)
analyzer.get_dictionary()
# 移除停止詞
analyzer.word_filter(w_len=0, # 過濾單詞字數 <= N的單詞
                     path_word='dictionary/stop_words.txt' # 停止詞字典
                    )
```

```python
# 載入 text monong 分析器
tm = text_mining(analyzer.split_list)
# get dictionary
tm.get_dictionary()
# 計算TF
tm.CounterVector()
# 計算TFIDF
tm.TfidfVector()
```

```python
All = []
for i in range(0, len(df.技術領域)):
    doc = []
    for j, _ in tm.TFIDF_Vector[i]:
        doc.append(tm.dic[j])
    All.append(doc)
```

```python
for i,j in enumerate(All):
    if len(j) == 0:
        All[i] = '無'
```

```python
from keras_bert import extract_embeddings

model_path = r'chinese_L-12_H-768_A-12'

Emb = extract_embeddings(model_path, df.技術領域)
```

```python
DF = pd.DataFrame({'斷詞內容':All, 'BERT_WordEmbedding':Emb, 'TFIDF':tm.TFIDF_Vector})
```

```python
DF['BERT_WordEmbedding_mean'] = DF.BERT_WordEmbedding.apply(lambda x: x.mean(axis=0))
```

```python
emb_sentence = DF['BERT_WordEmbedding_mean'].apply(lambda x: x.reshape(1,-1))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

sim = []
for i in range(0, len(df)):
    sim.append(cosine_similarity(emb_sentence[15], emb_sentence[i]))
```
